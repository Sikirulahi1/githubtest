# Seaborn: scatterplot
1. hue: Color change and category
2. Style: Marker(o,+,-, etc)
3. Size: Point style

```python
fig, ax = plt.subplots(2, 2, figsize=(10,10))

sns.scatterplot(data=data, x='trestbps', y='oldpeak', alpha=0.7, ax= ax[0][0])
sns.scatterplot(data=data, x='trestbps', y='oldpeak', alpha=0.7, hue='cp', ax= ax[0][1])

# The shade in graph 2 makes it difficult to differentiate between two parameters
# Instead of having shade, you can have different patterns
# The third graph differentiates the parameters in shapes
sns.scatterplot(data=data, x='trestbps', y='oldpeak',style='cp', color='royalblue', ax= ax[1][0])

# The fourth graph differentiates the parameters in sizes
sns.scatterplot(data=data, x='trestbps', y='oldpeak',size='cp', color='brown', ax= ax[1][1])
```

```python
# You can haave two different hue variable, where you have the differentiation according to size and hue
plt.figure(1, figsize=(6, 6))
sns.scatterplot(data=data, x='trestbps', y='oldpeak', style='sex', hue='target', color='royalblue', alpha=0.7)
```

# Seaborn: boxplot, stripplot, swarmplot

```python
fig, ax = plt.subplots(2, 2, figsize=(14,12))

# For boxplot
sns.boxplot(data=data, x='thal', y='thalach', ax=ax[0][0])

# For stripplot
sns.stripplot(data=data, x='thal', y='thalach', ax=ax[0][1], alpha=0.5)

# For swarmplot
sns.swarmplot(data=data,x='thal', y= 'thalach', ax=ax[1][0], alpha= 0.7)

# For both
sns.boxplot(data=data, x='thal', y='thalach', ax=ax[1][1])
sns.swarmplot(data=data, x='thal', y='thalach', ax=ax[1][1], alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# We can add hue parameter
fig, ax = plt.subplots(1, figsize=(12, 6))
sns.boxplot(data=data, x='thal', y='thalach', hue='target')
plt.show()
```

# Violin plot

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols= 3)

ax1.violinplot(data['trestbps'], showmedians=True, showmeans=True, vert=False)
ax1.set_title('resting bp')

ax2.violinplot(data['thalach'], showmedians=True, showmeans=True, vert=False)
ax2.set_title('Max heart rate')

ax3.violinplot(data['age'], showmedians=True, showmeans=True, vert=False)
ax3.set_title('age')
```

```python
fig, ax = plt.subplots(1, figsize=(6,6))

plt.title('age')
plt.violinplot(data['age'], showmedians=True, showmeans=True, vert=False)
plt.show()
```

# Seaborn: Violinplot

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot
sns.boxplot(x='slope', y='trestbps', data=data, ax=ax[0])

# Violinplot
sns.violinplot(x='slope', y='trestbps', data=data, ax=ax[1])
plt.show()
```

```python
# You can also apply the hue parameter

fig, ax = plt.subplots(1, 2, figsize=(14, 6))


sns.violinplot(x='slope', y='trestbps', data=data, hue='sex', ax=ax[0])
sns.violinplot(x='slope', y='trestbps', data=data, hue='target',split=True, ax=ax[1])
plt.show()
```

# Jointplot

```python
sns.set(style='white', color_codes=True)

sns.jointplot(data=data, x='age', y='oldpeak', alpha=0.7)
```

```python
# You can also set the kind of the presentation of the joint plot
#kind(reg,hex,kde)
sns.set(style='white', color_codes=True)
# reg means regression line
sns.jointplot(data=data, x='age', y='oldpeak', kind='reg')

# hex means hexagon 
sns.jointplot(data=data, x='age', y='oldpeak', kind='hex', color='gold')

# kde means it looks more like a contour
sns.jointplot(data=data, x='age', y='oldpeak', kind='kde', color='forestgreen')
```
