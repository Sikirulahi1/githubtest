# Finally, lets train the data

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=100000).fit(X_train, y_train)
```

```python
print("Training accuracy: {:.2f}".format(lr.score(X_train, y_train)))
```

```python
print("Test accuracy: {:.2f}".format(lr.score(X_test, y_test)))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth= 9).fit(X_train, y_train)
```

```python
print("Training accuracy: {:.2f}".format(tree.score(X_train, y_train)))
```

```python
print("Test accuracy: {:.2f}".format(tree.score(X_test, y_test)))
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB

bernoulli = BernoulliNB().fit(X_train, y_train)
gaussian = GaussianNB(var_smoothing= 0.1).fit(X_train, y_train)

# Checking the accuracy

print("Bernoulli Training accuracy: {:.2f}".format(bernoulli.score(X_train, y_train)))
print(" Bernoulli Test accuracy: {:.2f}".format(bernoulli.score(X_test, y_test)))

print("Gaussian Training accuracy: {:.2f}".format(gaussian.score(X_train, y_train)))
print("Gaussian Test accuracy: {:.2f}".format(gaussian.score(X_test, y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor = KNeighborsClassifier(n_neighbors= 2).fit(X_train, y_train)


# Checking the accuracy

print("Training accuracy: {:.2f}".format(neighbor.score(X_train, y_train)))
print("Test accuracy: {:.2f}".format(neighbor.score(X_test, y_test)))
```
